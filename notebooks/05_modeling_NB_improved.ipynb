```python
import pickle
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix

# 1. BUKA KEMBALI DATA YANG SUDAH JADI BINER (Critical vs Non-Critical)
print("Memuat data splits...")
with open('../data/processed/data_splits_tfidf.pkl', 'rb') as f:
    data = pickle.load(f)

X_train = data['X_train']
X_test = data['X_test']
y_train = data['y_train']
y_test = data['y_test']

# 2. INISIALISASI DAN LATIH MODEL NAIVE BAYES
print("Melatih model Multinomial Naive Bayes...\n")
model_nb = MultinomialNB()
model_nb.fit(X_train, y_train)

# 3. LAKUKAN PREDIKSI
y_pred_nb = model_nb.predict(X_test)

# 4. CETAK CLASSIFICATION REPORT
print("=== CLASSIFICATION REPORT (NAIVE BAYES) ===")
print(classification_report(y_test, y_pred_nb, target_names=['Critical', 'Non-Critical']))

# 5. CETAK CONFUSION MATRIX (FORMAT TEKS)
cm_nb = confusion_matrix(y_test, y_pred_nb)

print("\n=== CONFUSION MATRIX (NAIVE BAYES) ===")
print("                     [PREDIKSI MODEL]")
print("                    Critical    Non-Critical")
print(f"[ASLI] Critical    [{cm_nb[0][0]:<8}  {cm_nb[0][1]:<8}]")
print(f"[ASLI] Non-Crit    [{cm_nb[1][0]:<8}  {cm_nb[1][1]:<8}]")
```

```python
# Menyimpan model Naive Bayes
with open('../data/processed/model_nb_biner.pkl', 'wb') as f:
    pickle.dump(model_nb, f)
    
print("Model Naive Bayes berhasil disimpan untuk komparasi!")
```
