```python
import pickle
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report, confusion_matrix

# 1. LOAD DATA SPLITS
print("Memuat data splits...")
with open('../data/processed/data_splits_tfidf.pkl', 'rb') as f:
    data = pickle.load(f)

X_train = data['X_train']
X_test = data['X_test']
y_train = data['y_train']
y_test = data['y_test']

# 2. INISIALISASI DAN LATIH MODEL SVM

print("Melatih model Support Vector Machine (LinearSVC)...")
model_svm = LinearSVC(class_weight='balanced', random_state=42, max_iter=2000)
model_svm.fit(X_train, y_train)
print("Pelatihan selesai!\n")

# 3. LAKUKAN PREDIKSI
y_pred_svm = model_svm.predict(X_test)

# 4. CETAK CLASSIFICATION REPORT
print("=== CLASSIFICATION REPORT (SVM) ===")
print(classification_report(y_test, y_pred_svm, target_names=['Critical', 'Non-Critical']))

# 5. CETAK CONFUSION MATRIX
cm_svm = confusion_matrix(y_test, y_pred_svm)

print("\n=== CONFUSION MATRIX (SVM) ===")
print("                     [PREDIKSI MODEL]")
print("                    Critical    Non-Critical")
print(f"[ASLI] Critical    [{cm_svm[0][0]:<8}  {cm_svm[0][1]:<8}]")
print(f"[ASLI] Non-Crit    [{cm_svm[1][0]:<8}  {cm_svm[1][1]:<8}]")

# 6. SIMPAN MODEL SVM
with open('../data/processed/model_svm_biner.pkl', 'wb') as f:
    pickle.dump(model_svm, f)
    
print("\nModel SVM berhasil disimpan di model_svm_biner.pkl!")
```
